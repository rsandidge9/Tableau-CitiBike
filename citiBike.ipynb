```python
bike19_df= pd.concat(map(pd.read_csv, glob.glob('Data/19/*.csv')))
bike19_df
```

```python
bike19_df.info()
bike19_df['DateTime'] = pd.to_datetime(bike19_df['starttime'])
bike19_df['Date'] = bike19_df['DateTime'].dt.strftime('%d/%m/%Y')
bike19_df['Time'] = bike19_df['DateTime'].dt.strftime('%H:%M')
```

```python
bike19_df.to_csv('bike19.csv')
```
